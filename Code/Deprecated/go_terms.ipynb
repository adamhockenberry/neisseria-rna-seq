```python
keywords_dict = {}
goterms_dict = {}
for infile in glob.glob('../Data/Uniprot/*.dat')[:]:
    locus_name = ''
    handle = open(infile)
    record = SwissProt.read(handle)
    temp = record.gene_name.split(';')
    for i in temp:
        if 'OrderedLocusNames' in i:
            locus_name = i.split('=')[-1].strip()
    if locus_name == '':
        for i in temp:
            if 'ORFNames' in i:
                locus_name = i.split('{')[0].split('=')[-1].replace('_', '').strip()
    if locus_name != '':
        keywords_dict[locus_name] = [i.split('{')[0].strip() for i in record.keywords if i.find('}') == -1]
        temp_go = []
        for i in record.cross_references:
            if i[0] == 'GO':
                temp_go.append(i[2])
        goterms_dict[locus_name] = temp_go
```

```python
df1 = pd.read_csv('/Users/adamhockenberry/workspace/kallisto/neisseria_04_22_16/Neisseria_kallisto_test_table.csv', index_col='target_id')
df1 = df1.drop('Unnamed: 0', 1)
df2 = pd.read_csv('/Users/adamhockenberry/workspace/kallisto/neisseria_04_22_16/neisseria_kallisto_table.csv')
df2 = df2.drop('Unnamed: 0', 1)
```

```python
for i in ['SQ1', 'SQ2', 'SQ3', 'SQ4', 'SQ5', 'SQ6']:
    df1[i] = ''
```

```python
for index in df2.index:
    df1.ix[df2.loc[index]['target_id'], df2.loc[index]['sample']] = df2.loc[index]['tpm']
```

```python
df1['avg(SQ1,SQ3,SQ5)'] = np.nan
df1['avg(SQ2,SQ4,SQ6)'] = np.nan
df1['log2_FC'] = np.nan
for index in df1.index:
    try:
        df1.ix[index, 'avg(SQ2,SQ4,SQ6)'] = np.mean([df1.loc[index]['SQ2'], df1.loc[index]['SQ4'], df1.loc[index]['SQ6']])
    except TypeError:
        pass
    try:
        df1.ix[index, 'avg(SQ1,SQ3,SQ5)'] = np.mean([df1.loc[index]['SQ1'], df1.loc[index]['SQ3'], df1.loc[index]['SQ5']])
    except TypeError:
        pass
    try:
        df1.ix[index, 'log2_FC'] = np.log2(np.exp(df1.ix[index, 'b']))
    except TypeError:
        pass
```

```python
df_compact = df1[['SQ1', 'SQ2', 'SQ3', 'SQ4', 'SQ5', 'SQ6', 'avg(SQ1,SQ3,SQ5)', 'avg(SQ2,SQ4,SQ6)', 'log2_FC', 'qval', 'pval']]
```

```python
df_compact_sorted = df_compact.sort_index(axis=0)
```

```python
all_genes = list(df_compact.index)
print('Number of genes in Kallisto analysis: {}'.format(len(all_genes)))
diff_genes = list(df_compact[(df_compact.log2_FC > 0) & (df_compact.qval < 0.05)].index)
print('Number of differentially expressed genes in Kallisto analysis: {}'.format(len(diff_genes)))
```

```python
for goterm, n_genes in goterm_counter.items():
    total_n = 0
    total_condition_n = 0
    subset_n = 0
    subset_condition_n = 0
    if n_genes > 2:
        for i,j in goterms_dict.items():
            if i in all_genes:
                total_n += 1
                if goterm in goterms_dict[i]:
                    total_condition_n += 1
            if i in diff_genes:
                subset_n += 1
                if goterm in goterms_dict[i]:
                    subset_condition_n += 1
        test_stat = stats.hypergeom.sf(subset_condition_n-1, total_n, total_condition_n, subset_n)
        if test_stat < 0.01:
            print('{}\t{}\n{}\t{}\t{}\t{}\t{}'.format(goterm, n_genes, total_n, total_condition_n, subset_n, subset_condition_n, test_stat))
        test_stat = stats.hypergeom.cdf(subset_condition_n, total_n, total_condition_n, subset_n)
        if test_stat < 0.01:
            print('####{}\t{}\n{}\t{}\t{}\t{}\t{}'.format(goterm, n_genes, total_n, total_condition_n, subset_n, subset_condition_n, test_stat))
```

```python
for keyword, n_genes in keywords_counter.items():
    total_n = 0
    total_condition_n = 0
    subset_n = 0
    subset_condition_n = 0
    if n_genes > 2:
        for i,j in keywords_dict.items():
            if i in all_genes:
                total_n += 1
                if keyword in keywords_dict[i]:
                    total_condition_n += 1
            if i in diff_genes:
                subset_n += 1
                if keyword in keywords_dict[i]:
                    subset_condition_n += 1
        test_stat = stats.hypergeom.sf(subset_condition_n-1, total_n, total_condition_n, subset_n)
        if test_stat < 0.01:
            print('{}\t{}\n{}\t{}\t{}\t{}\t{}'.format(keyword, n_genes, total_n, total_condition_n, subset_n, subset_condition_n, test_stat))
        test_stat = stats.hypergeom.cdf(subset_condition_n, total_n, total_condition_n, subset_n)
        if test_stat < 0.01:
            print('####{}\t{}\n{}\t{}\t{}\t{}\t{}'.format(keyword, n_genes, total_n, total_condition_n, subset_n, subset_condition_n, test_stat))
```
